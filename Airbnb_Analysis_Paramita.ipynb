# **4. UNIVARIATE ANALYSIS**

**Distribution of Prices**

```python
sns.histplot(df['price'], bins=50, kde=True, color='green')
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price')
plt.show()
```

**Removing Outliers (Price IQR)**

```python
def remove_outliers_iqr(group_df):
    Q1 = group_df['price'].quantile(0.25)
    Q3 = group_df['price'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return group_df[(group_df['price'] >= lower_bound) & (group_df['price'] <= upper_bound)]

# Group-wise filtering and definition of df_filtered
df_filtered = df.groupby('room_type').apply(remove_outliers_iqr).reset_index(drop=True)
```

```python
sns.histplot(df_filtered['price'], bins=50, kde=True, color='blue')
plt.title('Distribution of Airbnb Prices')
plt.xlabel('Price')
plt.show()
```
